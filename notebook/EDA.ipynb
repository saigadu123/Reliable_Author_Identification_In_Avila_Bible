```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/saigadu123/Location-of-Avila-file/main/avila-tr.csv')
```

```python
df.columns = ["Intercolumnar_distance","upper_margin","lower_margin","exploitation","row_number","modular_ratio","inter_linear_spacing","weight","peak_number","modular_ratio/inter_linear_spacing","class"]
```

```python
for feature in df.columns:
    df[feature].hist()
    plt.show()
```

```python
for feature in df.columns[:-1]:
    lower_bound = df[feature].quantile(0.25)
    upper_bound = df[feature].quantile(0.50)
    #print("Feature : "+feature+"lower bound is ",lower_bound," and upper bound is ",upper_bound)
    IQR = upper_bound-lower_bound
    lower_boundary = lower_bound-(3*IQR)
    upper_boundary = upper_bound+(3*IQR)
    lower_index = df[df[feature]<lower_boundary].index
    upper_index = df[df[feature]>upper_boundary].index
    df.drop(lower_index,axis=0,inplace=True)
    df.drop(upper_index,axis=0,inplace=True)
    
```

```python
for feature in df.columns[:-1]:
    df[feature].hist()
    plt.show()
```
